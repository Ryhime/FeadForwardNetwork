```python
import pandas as pd
import numpy as np
ys = np.array(pd.read_csv("Data.csv")["price"])/7700000
```

```python
df = pd.read_csv("Data.csv")
rooms = np.array(df["bedrooms"])/33
baths = np.array(df["bathrooms"])/8
sqftLiving = np.array(df["sqft_living"])/13540
sqftLoft = np.array(df["sqft_lot"])/1651359
floors = np.array(df["floors"])/3.5
waterfront = np.array(df["waterfront"])
view = np.array(df["view"])/4
grade = np.array(df["grade"])/13
sqftAbove = np.array(df["sqft_above"])/9410
sqftBase = np.array(df["sqft_basement"])/4820
yearBuilt = np.array(df["yr_built"])/2015


print(min(yearBuilt))
xs = np.column_stack([rooms,baths,sqftLiving,sqftLoft,floors,waterfront,view,grade,sqftAbove,sqftBase])
```

```python
xsTrain = list(map(lambda x: np.matrix(x).T,xs))[0:20000]
ysTrain = list(map(lambda x: np.matrix(x).T,ys))[0:20000]

xsTest = list(map(lambda x: np.matrix(x).T,xs))[20000:21000]
ysTest = list(map(lambda x: np.matrix(x).T,ys))[20000:21000]
```

```python
from FeadForwardNetwork.FeedForward import Network
net = Network([10,5,1],learningRate=.01)
for i in range(20):
    net.epoch(xsTrain,ysTrain)
    print("Train:",net.computeLoss(xsTrain,ysTrain),"Test:",net.computeLoss(xsTest,ysTest))
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
model = LinearRegression()
model.fit(xs[0:20000],ys[0:20000])
```

```python
def loss(predicted,actual):
    return .5*(actual-predicted)**2
def computeLoss(preds,actuals):
    totalLoss = 0
    for i in range(len(preds)):
        totalLoss+=loss(preds[i],actuals[i])
    return totalLoss
```

```python
preds = model.predict(xs[20000:21000])
computeLoss(preds,ys[20000:21000])
```
